# Кластеризация ЕГЭ (4 если сделаны все задачи)

Рядом лежат данные с координатами точек. Везде используется Евклидово расстояние. Кластером считается набор не менее чем из 30 точек связанных друг с другом. Аномалия это точка находящаяся на расстоянии более 1 от любого кластера.

* Постройте Распределение точек
* Напишите руками DBSCAN и обработайте им все файлы
* Файл 0.xls также решите руками
* Постройте Распределение точек, отметьте принадлежность кластеров цветами
* Отметьте Аномалии отдельным цветом
* Найдите среди в каждом кластере точку расстояние от которой до всех остальных минимально
* Выведите два числа - Среднее абсцисс и ординат центроидов кластеров * 10000

```python
import matplotlib.pyplot as plt
import math

def eug_dist(x1, y1, x2, y2):
    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

points = []
with open("../0.txt", "r") as f:
    for line in f:
        line = line.replace(",", ".")
        points.append(list(map(float, line.split())))

eps = 1
min_pts = 30

visited = [False] * len(points)
cluster_labels = [-1] * len(points)
cluster_id = 0

def region_query(idx):
    neighbors = []
    for i, (x, y) in enumerate(points):
        if eug_dist(points[idx][0], points[idx][1], x, y) <= eps:
            neighbors.append(i)
    return neighbors

def expand_cluster(idx, neighbors):
    global cluster_id
    cluster_labels[idx] = cluster_id

    i = 0
    while i < len(neighbors):
        point_idx = neighbors[i]

        if not visited[point_idx]:
            visited[point_idx] = True
            new_neighbors = region_query(point_idx)
            if len(new_neighbors) >= min_pts:
                neighbors += new_neighbors

        if cluster_labels[point_idx] == -1:
            cluster_labels[point_idx] = cluster_id

        i += 1

for i in range(len(points)):
    if visited[i]:
        continue

    visited[i] = True
    neighbors = region_query(i)

    if len(neighbors) < min_pts:
        cluster_labels[i] = -1
    else:
        expand_cluster(i, neighbors)
        cluster_id += 1

print("Количество кластеров:", cluster_id)

clusters = {}
for idx, label in enumerate(cluster_labels):
    if label == -1:
        continue
    clusters.setdefault(label, []).append(idx)

medoids = []
for label, indices in clusters.items():
    min_sum = float("inf")
    best_point = None

    for i in indices:
        dist_sum = 0
        for j in indices:
            dist_sum += eug_dist(points[i][0], points[i][1],
                                 points[j][0], points[j][1])
        if dist_sum < min_sum:
            min_sum = dist_sum
            best_point = points[i]

    medoids.append(best_point)

centroids = []
for label, indices in clusters.items():
    x_avg = sum(points[i][0] for i in indices) / len(indices)
    y_avg = sum(points[i][1] for i in indices) / len(indices)
    centroids.append((x_avg, y_avg))

mean_x = sum(c[0] for c in centroids) / len(centroids)
mean_y = sum(c[1] for c in centroids) / len(centroids)

print(int(mean_x * 10000), int(mean_y * 10000))

colors = ['blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']

plt.figure(figsize=(8, 8))

for label, indices in clusters.items():
    xs = [points[i][0] for i in indices]
    ys = [points[i][1] for i in indices]
    plt.scatter(xs, ys, color=colors[label % len(colors)], s=10)

noise_x = [points[i][0] for i in range(len(points)) if cluster_labels[i] == -1]
noise_y = [points[i][1] for i in range(len(points)) if cluster_labels[i] == -1]
plt.scatter(noise_x, noise_y, color='red', s=15)

for mx, my in medoids:
    plt.scatter(mx, my, color='black', s=80, marker='x')

plt.grid(True)
plt.show()
```
